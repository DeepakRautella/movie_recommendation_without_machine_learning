```python
import pandas as pd
netflixdata=pd.read_csv('netflix_titles.csv')
netflixdata
```

```python
### for content based recommendation we will use title,cast,director,rating, listing in
```

```python
#drop unnecessary rows
nf=netflixdata.drop(['show_id','type','date_added','release_year','duration','description'],axis=1)
nf.head()
```

```python
# to fill nan with other substitue so that while adding rows nan dont add
```

```python
nf=nf.fillna('-')
nf.head()
```

```python
# Developing Recommendation Engine using cast, director, country, rating and listed_in
#combined all these feature into new row
```

```python
nf["combined_features"]=nf['director']+","+nf['cast']+","+nf['country']+","+nf['rating']+","+nf['listed_in']
```

```python
nf.head()
```

```python
# compare watched movies combined features data with remaining combined features and calculate % of recommendation
```

```python
def movie_recommendation(watched_movie):
    if watched_movie in list(nf['title']):
       
     # converting combined data one by one into list
                
            watched_movie_data=nf[nf['title']==watched_movie]
            combined_data=watched_movie_data['combined_features'].values.tolist() #watched movie combined data into list 
            combined_data=combined_data[0].split(",")                         #convert watched combined_features into list

            count_row=[]                       #matching data count storing 
            
            for k in nf['combined_features']:
                count=0                    #matching words count per combined data column
                data=list(k.split(','))     #converting movies combined features into list 

                for j in combined_data:
                    
                    if j in data:                            #count number of matchings of terms 
                        count=count+1
           
                count_row.append(count)
# convert count rows into percentage of movie recommendation
            matched_percent=[]
            for l in count_row:
                matched_percent.append((l/len(combined_data)*100))
            
#new movies databases after all operations
            nf['recommendation_per']=matched_percent
            new_data=nf[nf['title']!=watched_movie]       #drop same movie matched
            new_data=new_data.drop(['combined_features'],axis=1)       #dropped unneccasry row
            display(new_data.sort_values(by=['recommendation_per'],ascending=False).head())   
    else:
        print("Entered movie not available in database ")
        
        
```

```python
watched_movie=input("Enter watched movie :")
movie_recommendation(watched_movie)
```
